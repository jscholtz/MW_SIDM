```python
import numpy as np
from matplotlib import pyplot as plt
from tqdm.notebook import tqdm

from scipy.interpolate import interp1d

import contextlib
import joblib
from joblib import Parallel, delayed

@contextlib.contextmanager
def tqdm_joblib(tqdm_object):
    """Context manager to patch joblib to report into tqdm progress bar given as argument"""
    class TqdmBatchCompletionCallback(joblib.parallel.BatchCompletionCallBack):
        def __call__(self, *args, **kwargs):
            tqdm_object.update(n=self.batch_size)
            return super().__call__(*args, **kwargs)

    old_batch_callback = joblib.parallel.BatchCompletionCallBack
    joblib.parallel.BatchCompletionCallBack = TqdmBatchCompletionCallback
    try:
        yield tqdm_object
    finally:
        joblib.parallel.BatchCompletionCallBack = old_batch_callback
        tqdm_object.close()
```

# The goal of this part is to explore the multipole expansion method

in particular I need to figure out how to correctly incorporate symmetries on grids...

The full expression for the potential reads:

$$\Phi(\vec{R}) = \sum_l \sum_m \frac{1}{2l+1} \left[r^{l} P^*_{lm}(r) + r^{-l-1} Q^*_{lm}(r)\right] Y_{lm}(\theta,\phi)$$

where:

$$P_{lm}(r) = \int d\Omega \int_r^\infty x^2 dx \rho(x,\theta,\phi)Y_{lm}(\theta,\phi) x^{-l-1} $$

$$Q_{lm}(r) = \int d\Omega \int_0^r x^2 dx \rho(x,\theta,\phi)Y_{lm}(\theta,\phi) x^l$$


With the appropriate $Y_{lm}$'s

Stolen from: https://farside.ph.utexas.edu/teaching/jk1/Electromagnetism/node29.html (also see: https://en.wikipedia.org/wiki/Table_of_spherical_harmonics)

$\displaystyle Y_{0,0}(\theta,\varphi)$	$\displaystyle = \left(\frac{1}{4\pi}\right)^{1/2}$

$\displaystyle Y_{1,0}(\theta,\varphi)$	$\displaystyle = \left(\frac{3}{4\pi}\right)^{1/2}\cos\theta \to z/r,$

$\displaystyle Y_{2,0}(\theta,\varphi)$	$\displaystyle =\left(\frac{5}{16\pi}\right)^{1/2}(3\,\cos^2\theta-1) \to (3z^2-r^2)/r^2$



```python
G_Newton = 4.5e-6
par_dm = {"rho0": 8.54e6, "rh": 19.6}
par_blg = {"alpha": 1.8, "q": 0.5, "r0": 0.075, "rcut": 2.10, "rho0":  9.93e10}
par_sd = {"Rd_thin": 2.5, "Rd_thick": 3.0, "Sigma0_thin": 8.92e8, "Sigma0_thick": 1.83e8,
          "zd_thin": 0.3, "zd_thick": 0.9}
par_gd = {"Sigma0_HI": 5.30e7, "Sigma0_H2": 2.18e9, "Rd_HI": 7.0, "Rd_H2": 1.5,
          "Rm_HI": 4.0, "Rm_H2": 12.0, "zd_HI": 0.085, "zd_H2": 0.045}
```

```python
def pot_NFW(R):
    
    Rn = R/par_dm['rh']
    return -4*np.pi*G_Newton*par_dm['rh']**2*par_dm['rho0']*( (np.log(1.0+Rn)+1.0/(1.0+Rn)-1.0)/Rn + 1.0/(1.0+Rn) )
```

```python
def make_props(nsides,boundary):

    rs = np.linspace(0,boundary,nsides)
    zs = np.linspace(0,boundary,nsides)
    dx = rs[1]

    rr,zz = np.meshgrid(rs,zs,indexing='ij')
    RR = np.sqrt(rr**2+zz**2)

    measure = np.copy(4.0*np.pi*rr*dx**2)
    measure[:,0] = measure[:,0]/2.0 

    RR[0,0] = dx/4
    Rs = RR[0]

    mask_spherical = (RR > boundary)

    RR_san = np.copy(RR)
    RR_san[mask_spherical] = boundary

    props = {'nsides': nsides, 'boundary': boundary, 'mask_spherical': mask_spherical,
             'rs': rs, 'zs': zs, 'dx': dx, 'rr': rr, 'zz': zz, 'RR': RR,
             'measure': measure, 'Rs': Rs, 'RR_san': RR_san, 'costh': zz/RR }
    
    return props
```

```python
def construct_NFW(props):
    RR = props['RR']
    boundary = props['boundary']
    mask = props['mask_spherical']
    
    density = par_dm["rho0"] * par_dm["rh"]/(RR*(1.0+RR/par_dm["rh"])**2)
    density[mask] = 0
    
    return density


def construct_gal(props):
    
    boundary = props['boundary']
    rr = props['rr']
    zz = props['zz']
    dx = props['dx']
    mask = props['mask_spherical']
    
    r_prime = np.sqrt(rr**2 + (zz/par_blg["q"])**2)

    density = par_blg["rho0"] * np.exp(-(r_prime/par_blg["rcut"])**2) / np.power((1.0+r_prime/par_blg["r0"]), par_blg["alpha"])
    density = density + par_sd["Sigma0_thin"]/(2.0*par_sd["zd_thin"]) * np.exp(-np.abs(zz)/par_sd["zd_thin"] - rr/par_sd["Rd_thin"])
    density = density + par_sd["Sigma0_thick"]/(2.0*par_sd["zd_thick"]) * np.exp(-np.abs(zz)/par_sd["zd_thick"] - rr/par_sd["Rd_thick"])
    
    ## KEEP IN MIND THE BELLOW IS REGULARISED BY HAND!!!
    
    mask1 = (zz/(2.0*par_gd["zd_HI"]) <= 100)
    temp2 = np.zeros_like(zz)
    temp2[mask1] = 1/np.cosh(zz[mask1]/(2.0*par_gd["zd_HI"]))**2
    density = density + par_gd["Sigma0_HI"]/(4.0*par_gd["zd_HI"]) * np.exp(- par_gd["Rm_HI"]/np.sqrt(rr**2+dx**2) - rr/par_gd["Rd_HI"])*temp2
    
    mask1 = (zz/(2.0*par_gd["zd_H2"]) <= 100)
    temp2 = np.zeros_like(zz)
    temp2[mask1] = 1/np.cosh(zz[mask1]/(2.0*par_gd["zd_H2"]))**2
    density = density + par_gd["Sigma0_H2"]/(4.0*par_gd["zd_H2"]) * np.exp(- par_gd["Rm_H2"]/np.sqrt(rr**2+dx**2) - rr/par_gd["Rd_H2"])*temp2
    
    density[mask] = 0
    
    return density
```

```python
def get_PQs_serial(full_den,Rs,props):

    RR = props['RR']
    zz = props['zz']
    measure = props['measure']
    
    P00 = []
    Q00 = []

    P20 = []
    Q20 = []

    Q00_temp = measure*full_den
    P00_temp = measure*full_den/(RR)

    Q20_temp = measure*(3*zz**2-RR**2)*full_den      
    P20_temp = measure*(3*zz**2-RR**2)*full_den/(RR**5)        
    
    for R in tqdm(Rs):
        mask_in =  ( RR <= R )
        mask_out = ( RR > R )

        P00.append(np.sum(P00_temp[mask_out]))
        Q00.append(np.sum(Q00_temp[mask_in]))

        P20.append(np.sum(P20_temp[mask_out]))
        Q20.append(np.sum(Q20_temp[mask_in]))
        
    return np.array(P00),np.array(Q00),np.array(P20),np.array(Q20)
```

```python
def get_C(PQs_NFW,Rs):

    Phi_NFW_00 = 4.0*np.pi*G_Newton * ((PQs_NFW[1]/Rs + PQs_NFW[0])*(1/(4.0*np.pi)))
    C = -(Phi_NFW_00[-1]+pot_NFW(Rs)[-1])
    return C
```

```python
def get_PQs(full_den,Rs,props):

    RR = props['RR']
    zz = props['zz']
    measure = props['measure']
    
    Q00_int = measure*full_den
    P00_int = measure*full_den/(RR)

    Q20_int = measure*(3*zz**2-RR**2)*full_den      
    P20_int = measure*(3*zz**2-RR**2)*full_den/(RR**5)        
    
    def calc(Q00_int,P00_int,Q20_int,P20_int,RR,R):
        mask_in =  ( RR <= R )
        mask_out = ( RR > R )
        return [np.sum(P00_int[mask_out]),np.sum(Q00_int[mask_in]),np.sum(P20_int[mask_out]),np.sum(Q20_int[mask_in])]
    
    with tqdm_joblib(tqdm(desc="Calc", total=len(Rs))) as progress_bar:
        ans = Parallel(n_jobs=4)(delayed(calc)(Q00_int,P00_int,Q20_int,P20_int,RR,R) for R in Rs)
        
    return np.array(ans).T
```

```python
def get_Phi(PQs,Rs,props):
    
    RR_san = props['RR_san']
    costh = props['costh']
    
    f00 = interp1d(Rs,PQs[1]/Rs    + PQs[0])
    f20 = interp1d(Rs,PQs[3]/Rs**3 + PQs[2]*Rs**2)

    return 4.0*np.pi*G_Newton * (f00(RR_san)*(1/(4.0*np.pi))+(1.0/5.0)*(3.0*costh**2-1.0)*f20(RR_san)*(5.0/(16.0*np.pi)))
```

```python
props = make_props(2**10+1,40.0)

NFW_den = construct_NFW(props)
Gal_den = construct_gal(props)
full_den = NFW_den + Gal_den

Rs = props['Rs']

PQs_NFW = get_PQs(NFW_den,Rs,props)
PQs_Gal = get_PQs(Gal_den,Rs,props)

Phi_full_NFW = get_Phi(PQs_NFW,Rs,props)
Phi_full_Gal = get_Phi(PQs_Gal,Rs,props)

Phi_full = Phi_full_NFW + Phi_full_Gal
```

```python
plt.figure(figsize=(15,10))

ax = plt.subplot(231,aspect='equal')
CS = plt.contour(props['rr'],props['zz'],np.log(NFW_den))
ax.clabel(CS, inline=True, fontsize=10)
ax.grid(linestyle='--')
ax.minorticks_on()

ax = plt.subplot(232,aspect='equal')
CS = plt.contour(props['rr'],props['zz'],np.log(Gal_den))
ax.clabel(CS, inline=True, fontsize=10)
ax.grid(linestyle='--')
ax.minorticks_on()

ax = plt.subplot(233,aspect='equal')
CS = plt.contour(props['rr'],props['zz'],np.log(full_den))
ax.clabel(CS, inline=True, fontsize=10)
ax.grid(linestyle='--')
ax.minorticks_on()

ax = plt.subplot(234,aspect='equal')
CS = plt.contour(props['rr'],props['zz'],np.log(Phi_full_NFW))
ax.clabel(CS, inline=True, fontsize=10)
ax.grid(linestyle='--')
ax.minorticks_on()

ax = plt.subplot(235,aspect='equal')
CS = plt.contour(props['rr'],props['zz'],np.log(Phi_full_Gal))
ax.clabel(CS, inline=True, fontsize=10)
ax.grid(linestyle='--')
ax.minorticks_on()

ax = plt.subplot(236,aspect='equal')
CS = plt.contour(props['rr'],props['zz'],np.log(Phi_full))
ax.clabel(CS, inline=True, fontsize=10)
ax.grid(linestyle='--')
ax.minorticks_on()

plt.show()
```

## Try the cycle of SIDM evolution

```python
props = make_props(2**10+1,40.0)

den_dm_0 = construct_NFW(props)
den_bar = construct_gal(props)
den_all_0 = den_dm_0 + den_bar

Rs = props['Rs']

PQs_dm_0 = get_PQs(den_dm_0,Rs,props)
PQs_bar =  get_PQs(den_bar, Rs,props)

Phi_dm_0 = get_Phi(PQs_dm_0,Rs,props)
Phi_bar  = get_Phi(PQs_bar, Rs,props)

Phi_all_0  = Phi_dm_0 + Phi_bar
```

```python
som = 2.1e-10 ## 1cm^2/g = 2.1e-10 kpc^2/M_sun
age = 10


C = get_C(PQs_dm_0,Rs)
print(C)

rate = som*den_dm_0*np.sqrt(C+Phi_all_0)

mask_inside  = (age*rate >= 3)
mask_outside = (age*rate < 3 )

Mcon = np.sum((props['measure']*den_dm_0)[mask_inside])
Econ = np.sum((props['measure']*den_dm_0*(-C+Phi_all_0))[mask_inside])

print(f"{Mcon=:.2e}")
print(f"{Econ=:.2e}")

v = np.sqrt(2*Econ/Mcon)
```

```python
v=185
den_dm_1 = np.exp(Phi_all_0/v**2)
Msum_1 = np.sum((props['measure']*den_dm_1)[mask_inside])
den_dm_1 = (Mcon/Msum_1)*den_dm_1
Msum_1 = np.sum((props['measure']*den_dm_1)[mask_inside])
den_dm_1[mask_outside] = den_dm_0[mask_outside]
Esum = 0.5*v**2*Msum_1
print(Esum/Econ)
```

```python
PQs_dm_1 = get_PQs(den_dm_1,Rs,props)
Phi_dm_1 = get_Phi(PQs_dm_1,Rs,props)
Phi_all_1  = Phi_dm_1 + Phi_bar
```

```python
v=185
den_dm_2 = np.exp(Phi_all_1/v**2)
Msum_2 = np.sum((props['measure']*den_dm_2)[mask_inside])
den_dm_2 = (Mcon/Msum_2)*den_dm_2
Msum_2 = np.sum((props['measure']*den_dm_2)[mask_inside])
den_dm_2[mask_outside] = den_dm_0[mask_outside]
Esum = 0.5*v**2*Msum_2
print(Esum/Econ)
```

```python
PQs_dm_2 = get_PQs(den_dm_2,Rs,props)
Phi_dm_2 = get_Phi(PQs_dm_2,Rs,props)
Phi_all_2  = Phi_dm_2 + Phi_bar
```

```python
v=185
den_dm_3 = np.exp(Phi_all_2/v**2)
Msum_3 = np.sum((props['measure']*den_dm_3)[mask_inside])
den_dm_3 = (Mcon/Msum_3)*den_dm_3
Msum_3 = np.sum((props['measure']*den_dm_3)[mask_inside])
den_dm_3[mask_outside] = den_dm_0[mask_outside]
Esum = 0.5*v**2*Msum_3
print(Esum/Econ)
```

```python
ax = plt.subplot(111,aspect='equal')
CS1 = plt.contour(props['rr'],props['zz'],np.log10(den_dm_0),levels=[6.0,6.2,6.4,6.6,6.8,7.0,7.2])
CS2 = plt.contour(props['rr'],props['zz'],np.log10(den_dm_1),levels=[6.0,6.2,6.4,6.6,6.8,7.0,7.2])
CS3 = plt.contour(props['rr'],props['zz'],np.log10(den_dm_2),levels=[6.0,6.2,6.4,6.6,6.8,7.0,7.2])
CS4 = plt.contour(props['rr'],props['zz'],np.log10(den_dm_3),levels=[6.0,6.2,6.4,6.6,6.8,7.0,7.2])

ax.clabel(CS1, fontsize=6)
ax.clabel(CS2, fontsize=6)
ax.clabel(CS3, fontsize=6)
ax.clabel(CS4, fontsize=6)

#ax.grid(linestyle='--')
ax.minorticks_on()
plt.xlim(0,15)
plt.ylim(0,15)
```
